```python
data_time = []
longitude, latitude = [] , []
magnitude, depth = [] , []
with open('data/quakes.csv') as fileToRead:
    for line in fileToRead.readlines():
        values= line.split(',')
        data_time.append(values[1])
        longitude.append(values[2])
        latitude.append(values[3])
        magnitude.append(values[4])
        depth.append(values[5])
```

```python
len(data_time), len(longitude), len(latitude), len(magnitude), len(depth)
```

```python
all_quakes_data = {
    'data_time' : data_time,
    'longitude' : longitude,
    'latitude' : latitude,
    'magnitude' : magnitude,
    'depth' : depth,
    }
print(all_quakes_data)
```
